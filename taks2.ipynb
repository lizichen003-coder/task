```python
import numpy as numpy
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
filename = r'C:\Users\bruce\Desktop\task\task2.png'
img = cv.imread(filename)
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
import numpy as np


gray = np.float32(gray)
dst = cv.cornerHarris(gray,2,3,0.04)
dst = cv.dilate(dst,None)
```

```python
img[dst > 0.01 * dst.max()] = [0, 0, 255]

# 显示结果
cv.imshow('Harris Corner Detection', img)

# 等待按键，按ESC键退出
if cv.waitKey(0) & 0xff == 27:
    cv.destroyAllWindows()
```

```python
# 导入必要的库
import cv2

# 读取输入图像
img = cv2.imread('task2.jpg')

# 将图像转换为灰度图
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 初始化SIFT对象
sift = cv2.SIFT_create()

# 检测关键点
keypoints = sift.detect(gray, None)

# 在图像上绘制关键点
output_img = cv2.drawKeypoints(gray, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# 显示结果
cv2.imshow("SIFT Keypoints", output_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
