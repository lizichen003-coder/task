```python
import cv2
import numpy as np
from IPython.display import display, Image


img = cv2.imread(r"E:\test\03785.png", 0)
x = cv2.Sobel(img, cv2.CV_16S, 1, 0)
y = cv2.Sobel(img, cv2.CV_16S, 0, 1)
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
dst = cv2.addWeighted(absX, 0.5, absY, 0.5, 0)


def show_in_notebook(image, title):
    _, encoded = cv2.imencode('.png', image)
    display(Image(data=encoded.tobytes()))
    print(title)

show_in_notebook(absX, "absX - X")
show_in_notebook(absY, "absY - Y")
show_in_notebook(dst, "Result -res")
```

Canny算子

```python
import cv2
import numpy as np
from IPython.display import display, Image

img = cv2.imread(r"E:\test\03785.png", 0)

if img is None:
    print("错误：无法加载图像")
else:
    
    edges = cv2.Canny(img, 50, 150)
    
   
    if img.shape[1] > 800:
        img = cv2.resize(img, (800, int(img.shape[0] * 800 / img.shape[1])))
        edges = cv2.resize(edges, (800, int(edges.shape[0] * 800 / edges.shape[1])))
    

    print("原图:")
    _, img_encoded = cv2.imencode('.png', img)
    display(Image(data=img_encoded.tobytes()))
    
    print("Canny边缘检测结果:")
    _, edges_encoded = cv2.imencode('.png', edges)
    display(Image(data=edges_encoded.tobytes()))
```

拉普拉斯

```python
import cv2
import numpy as np
from IPython.display import display, Image


img = cv2.imread(r"E:\test\03785.png", 0)  

if img is None:
    print("错误：无法加载图像")
else:
   
    blurred = cv2.GaussianBlur(img, (3, 3), 0)
    
   
    laplacian = cv2.Laplacian(blurred, cv2.CV_64F)
    
  
    laplacian_abs = cv2.convertScaleAbs(laplacian)
    
    print("原图:")
    _, img_encoded = cv2.imencode('.png', img)
    display(Image(data=img_encoded.tobytes()))
    
    print("拉普拉斯边缘检测结果:")
    _, laplacian_encoded = cv2.imencode('.png', laplacian_abs)
    display(Image(data=laplacian_encoded.tobytes()))
```
